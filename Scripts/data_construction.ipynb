```python
import pandas as pd
products_df = pd.read_csv("/content/product_info.csv")
```

```python
Skincare_df = products_df[products_df['primary_category'] == "Skincare"]
```

```python
# Define the list of secondary categories to filter
secondary_categories = [
    'Sunscreen',
    'Cleansers',
    'Moisturizers'
]

# Filter the DataFrame for the specified secondary categories
Skincare_df = Skincare_df[Skincare_df['secondary_category'].isin(secondary_categories)]
```

```python
# Filter out rows with no reviews (assuming the reviews are in a column named 'reviews')
Skincare_df  = Skincare_df [Skincare_df ['reviews'].notnull()]
```

```python
# importing review data sets
rev1 = pd.read_csv("/content/reviews_0-250.csv", low_memory=False)
rev2 = pd.read_csv("/content/reviews_250-500.csv", low_memory=False)
rev3 = pd.read_csv("/content/reviews_500-750.csv", low_memory=False)
rev4 = pd.read_csv("/content/reviews_750-1250.csv", low_memory=False)
rev5 = pd.read_csv("/content/reviews_1250-end.csv", low_memory=False)
```

```python
#filtering all of the review datasets for the columns we want
selected_columns = [
    'product_id',
    'author_id',
    'submission_time',
    'review_text',
    'review_title',
    'product_name',
    'brand_name',
    'price_usd'
]
rev1 = rev1[selected_columns]
rev2 = rev2[selected_columns]
rev3 = rev3[selected_columns]
rev4 = rev4[selected_columns]
rev5 = rev5[selected_columns]

# merging the 5 review datasets into one
revs = pd.concat([rev1, rev2, rev3, rev4, rev5])
print(revs.info())
```

```python
#merging review data set with the product dataset by common columns
Skincare_df = Skincare_df.merge(revs, left_on=['product_id','product_name','brand_name','price_usd'],
                       right_on=['product_id','product_name','brand_name','price_usd'])
Skincare_df.info()
```
